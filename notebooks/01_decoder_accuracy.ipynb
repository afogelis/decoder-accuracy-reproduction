# How Far Is MWPM From Optimal?

Reproducing arXiv:2311.12503. On small code-capacity surface codes we enumerate every error pattern
to compute the *exact* logical error rate of MWPM and of the optimal maximum-likelihood decoder.
MWPM is exactly optimal at d=3 and develops a small, growing sub-optimality at d=5.

```python
%matplotlib inline
import matplotlib.pyplot as plt

from darepro.accuracy import sweep_accuracy
from darepro.figures import plot_ler_vs_p, plot_suboptimality

results = sweep_accuracy(distances=[3, 5], error_rates=[0.02, 0.06, 0.10, 0.15])
for r in results:
    print(f"d={r.distance} p={r.physical_error_rate:.2f} "
          f"optimal={r.optimal_ler:.4e} mwpm={r.mwpm_ler:.4e} ratio={r.suboptimality_ratio:.4f}")
```

```python
plot_ler_vs_p(results)
plt.tight_layout()
plt.show()

plot_suboptimality(results)
plt.tight_layout()
plt.show()
```
